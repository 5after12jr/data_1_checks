```python
df = pd.read_csv('Louisville, Kentucky Weather 2020-07-12 to 2022-07-12.csv')
```

```python
pd.set_option('display.max_columns', None)
"""This uncovers the hidden columns in my dataframe"""
```

```python
df
```

```python
df.drop(df.loc[:, 'precip':'sealevelpressure'].columns, inplace=True, axis=1)
"""Drops precip, sealevelpressure, and all the columns inbetween them"""
```

```python
df.drop(df.loc[:, 'solarradiation':'conditions'].columns, inplace=True, axis=1)
"""Drops solarradiation, conditions, and all the columns inbetween them"""
```

```python
df.drop(['icon', 'stations'], inplace=True, axis=1)
"""Drops the columns icon and stations"""
```

```python
df.rename(columns = {'name':'Location', 'datetime':'Date', 'tempmax':'Max_Temperature', 'tempmin':'Min_Temperature', 'temp':'Temperature', 'feelslikemax':'Feels_Like_Max', 'feelslikemin':'Feels-Like_Min', 'feelslike':'Feels_Like', 'dew':'Dew', 'humidity':'Humidity', 'cloudcover':'Cloud_Cover', 'visibility':'Visibility'}, inplace = True)
"""Renaming all columns so they look better when visualized"""
```

```python
def avgMin_Temperature():
    ''' adds every value in Min_Temperature to create a sum
divides the sum by the number of values in mintemp
This will provide the average of Min_Temperature.
'''
    print(sum(df.Min_Temperature)/len(df.Min_Temperature))
"""Finding and then printing the average of Min_Temperature when called"""
```

```python
def avgMax_Temperature():
    ''' adds every value in Max_Temperature to create a sum
divides the sum by the number of values in maxtemp
This will provide the average of Max_Temperature.
'''
    print(sum(df.Max_Temperature)/len(df.Max_Temperature))

"""Finding and then printing the average of Max_Temperature when called"""
```
