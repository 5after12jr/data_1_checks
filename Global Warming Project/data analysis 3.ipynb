```python
df = pd.read_csv('Louisville, Kentucky Weather 2020-07-12 to 2022-07-12.csv')
```

```python
df.drop(df.loc[:, 'precip':'sealevelpressure'].columns, inplace=True, axis=1)
```

```python
df.drop(df.loc[:, 'visibility':'conditions'].columns, inplace=True, axis=1)
```

```python
df.drop(['icon', 'stations'], inplace=True, axis=1)
```

```python
df.rename(columns = {'name':'Location', 'datetime':'Date', 'tempmax':'Max Temperature', 'tempmin':'Min Temperature', 'temp':'Temperature', 'feelslikemax':'Feels Like Max', 'feelslikemin':'Feels Like Min', 'feelslike':'Feels Like', 'dew':'Dew', 'humidity':'Humidity', 'cloudcover':'Cloud Cover', 'visibility':'Visibility'}, inplace = True)
```

```python
def avgmintemp():
    ''' adds every value in mintemp to create a sum
divides the sum by the number of values in mintemp
This will provide the average of maxtemp.
'''
    print(sum(df.tempmin)/len(df.tempmin))
```

```python
def avgmaxtemp():
    ''' adds every value in maxtemp to create a sum
divides the sum by the number of values in maxtemp
This will provide the average of maxtemp.
'''
    print(sum(df.tempmax)/len(df.tempmax))
```
